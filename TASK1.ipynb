```python
import matplotlib.pyplot as plt
import numpy as np

# Sample population data for selected countries across different years
countries = ['USA', 'India', 'China', 'Brazil', 'Germany', 'Nigeria', 'UK', 'Japan']
years = ['2020', '2021', '2022']

# Hypothetical gender distribution data in millions
data = {
    'USA': {'male': [164, 165, 166], 'female': [167, 168, 169]},
    'India': {'male': [718, 720, 722], 'female': [662, 663, 665]},
    'China': {'male': [700, 699, 698], 'female': [688, 687, 686]},
    'Brazil': {'male': [106, 107, 108], 'female': [107, 108, 109]},
    'Germany': {'male': [41, 41, 42], 'female': [43, 43, 43]},
    'Nigeria': {'male': [104, 105, 106], 'female': [102, 103, 104]},
    'UK': {'male': [33, 33, 34], 'female': [34, 34, 35]},
    'Japan': {'male': [62, 62, 61], 'female': [55, 55, 54]}
}

# Create a bar chart
bar_width = 0.35  # Width of bars
x_labels = np.arange(len(countries))  # Label positions

plt.figure(figsize=(14, 8))

# Define colors for each year
colors = ['#007BFF', '#E83E8C', '#20C997']  # Blue, Pink, Teal

# Plotting bars for each year
for i, year in enumerate(years):
    male_population = [data[country]['male'][i] for country in countries]
    female_population = [data[country]['female'][i] for country in countries]
    
    # Positions for each bar group
    male_bars = plt.bar(x_labels - bar_width/2 + i * bar_width, male_population, bar_width, 
                        label=f'Male {year}', color=colors[i], alpha=0.7)
    female_bars = plt.bar(x_labels + bar_width/2 + i * bar_width, female_population, bar_width, 
                          label=f'Female {year}', color=colors[i], alpha=0.4)

# Add data labels on top of each bar
for bars in [male_bars, female_bars]:
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f'{yval}M', ha='center', va='bottom', fontsize=9)

# Update title and labels
plt.title('Population Distribution by Gender Across Selected Countries (2020-2022)', fontsize=16)
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Population (millions)', fontsize=12)
plt.xticks(x_labels, countries)  # Set the x-tick labels to the countries
plt.legend()
plt.grid(axis='y', alpha=0.75)

# Show the plot
plt.tight_layout()
plt.show()
```
